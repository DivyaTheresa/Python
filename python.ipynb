**EXPERIMENT 8 - develop A PYTHON PROGRAM TO CONVERT BINARY TO DECIMAL NUMBER AND VERIFY ITS PERFECT NUMBER**

eg:6 - DIVERSORS - 1,2,3--1+2+3=6--Its perfect
eg:15--1,3,5--1+3+5=9--not a pefect number

```python
binary=input('enter the binary number')
decimal=eval(binary)  #converts binary to decimal
print(decimal)
sum=0
count=1
while count<decimal:
  if decimal%count==0:
    sum=sum+count
    count+=1
  if sum==decimal:
    print('the given number is perfect number')
  else:
    print('the given number is not a perfect number')
```

SINE SERIES

```python
import math

def factorial(n):
  if n<2:
    return 1
    return n*factorial(n-1)

def sin(x, n):
  sine = 0
  for term in range(n):
      sign = (-1)**term
      y = x**(math.pi/180)
      sine +=  (((y**(2.0*term+1))/factorial(2*term+1))*sign)
      return sine

print(sin(30,5))
```

NUMBER to word conversion

```python
one= ["", "one", "two ", "three ", "four ","five ", "six", "seven ", "eight ", "nine ", "ten ", "eleven", "twelve ","thirteen ", "fourteen ", "fifteen ",
      "sixteen ", "seventeen ", "eighteen ","nineteen "]
ten = ["", "", "twenty ", "thirty ", "forty ","fifty ", "sixty ", "seventy ", "eighty ","ninety "]

def num_Words (n, s):
    str = ""
    if (n> 19):
        str += ten[n // 10] + one [n % 10]
    else:
        str += one[n]
    if (n):
        str += s
    return str

def convert_Words (n):
    out = "" # empty string
    out + num_Words ((n // 10000000),"crore ")
    out += num_Words (((n // 100000) %100),"lakh ")
    out += num_Words (((n // 1000) % 100),"thousand ")
    out + num_Words (((n // 100) % 10), "hundred ")
    if (n> 100 and n % 100):
        out += "and"
    out += num_Words ((n % 100), "")
    return out

n = 1590
print(convert_Words (n))
```
