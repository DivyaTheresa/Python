```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'Product': ['Product A', 'Product B', 'Product C', 'Product D'],
        'Sales': [25000, 18000, 32000, 12000]}
df = pd.DataFrame(data)
total_sales = df['Sales'].sum()
df['Percentage'] = (df['Sales'] / total_sales) * 100
plt.pie(df['Percentage'], labels=df['Product'], autopct='%1.1f%%')
plt.title('Total Sales for Last Year by Product')
plt.legend()
plt.show()
```

```python
#7.
import matplotlib.pyplot as plt

# Define the table of contents as a list of dictionaries
t= [
    {'Name': 'John Doe', 'Age': 20, 'Phone Number': '555-1234'},
    {'Name': 'Jane Smith', 'Age': 3, 'Phone Number': '555-5678'},
    {'Name': 'Bob Johnson', 'Age': 91, 'Phone Number': '555-9012'},
    {'Name': 'Alice Williams', 'Age': 118, 'Phone Number': '555-3456'},
    {'Name': 'David Brown', 'Age': 70, 'Phone Number': '555-7890'}
]

# Define the age ranges
age_ranges = [0, 20, 30, 40, 50, 60, 70,80,90,100,120,140]

# Extract the ages from the table of contents
ages = [person['Age'] for person in t]

# Create a histogram
plt.hist(ages, bins=age_ranges)

# Set the chart title and axis labels
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of People')

# Show the chart
plt.show()
```
