FOR DRIVER'S GENDER

```python
print(df.driver_gender.value_counts())
a=df.driver_gender.value_counts()

plt.pie(a)
mylabels=["Male","female"]
myexplode = [0,0.2]
plt.pie(a, labels=mylabels,explode=myexplode)
plt.show()

plt.plot(a,c="r",marker='o', markerfacecolor='blue')
plt.show()

a.plot.bar()

```
